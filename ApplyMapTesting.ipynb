```python
import pandas as pd
import numpy as np
print("Pandas Version: " + str(pd.__version__))
print("NumPy Version: " + str(np.__version__))
```

Creating an array of 100 numbers, arranged into a 10x10 matrix

```python
test_array = pd.DataFrame(np.array(range(100)).reshape(10,10))
```

```python
test_array
```

Default for the apply method is to _apply_ a function across a column. Shown below, the `np.sum` function is applied to each column, adding up all the values in each column and returning the sum.

```python
test_array.apply(np.sum)
```

Changing the kwarg axis to 1, means that the sum is applied to each row instead.

```python
test_array.apply(np.sum,axis=1)
```

```python
test_array.apply(np.mean,axis=1)
```

Difference in the three methods:
- `.apply()` is row- or column-wise.
- `.applymap()` is per element across a DataFrame
- `.map()` is per element across a Series

```python
test_array.apply(np.max)
```

```python
test_array.applymap(np.max)
```

```python
np.mod(0,3)
```

```python
pd.concat([test_array[0],test_array[0].map(np.max),test_array[0].apply(np.max)],axis=1)
```
